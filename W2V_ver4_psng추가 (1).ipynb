```python
### 업종들 categs로 분류
categs = ['운송','항공','해운','디스플레이','손해보험','가정용품','국방','건강관리','무역','소매','소프트웨어','출판','화학','게임','엔터테인먼트','IT','섬유','의류','가구','광고','철강','제약','에너지','기계','생물공학','목재','레저','카드','자동차','호텔','조선','부동산','건축','화장품']
```

```python
### 키워드 1위가 '반도체'라고 치고, 100개의 유사한 단어를 추출해 words에 저장
words = model.wv.most_similar("반도체", topn = 100)
```

```python
### 100개의 단어 모두 카테고리 별 유사도 계산하여 res에 저장
res = []
for word in words:
  a = []
  w = word[0] ### w2v 결과로 나온 단어 하나
  for categ in categs:
    try:
      ### 그 단어와 종목 카테고리 유사도 계산
      num = model.wv.similarity(w,categ) ### -> 단어와 유사도 높은 categs 나옴
      ### a에 유사도와 카테고리 저장
      a.append((num, categ)) 
      a.sort()
    except:
      continue

  ### b에 단어, positive, 카테고리 저장
  res.append((w, a))
```

### positive, negative 활용한 W2V 함수 (psng)
- 키워드 중 종목명이 아닌 것과 관련된 추천 종목을 찾기 위함
- **input** : 종목명이 아닌 키워드 (이하 '키워드')
- **logic**
  1. model.wv.most_similar로 키워드와 유사한 단어(이하 '주변 단어') 100개를 뽑음 
  
     (종목이 나올 가능성을 높게 하기 위해 키워드 하나 뿐만 아니라 그와 유사한 단어들에도 워드투백 시행)
  2. 100개의 주변 단어에 대해 각각 wv.most_similar(positive=['키워드','급등'], topn=10) 으로 10개의 연관 단어를 뽑음
- **output** : (주변단어, [해당 단어에 대해 종목과 관련된 단어들'] ) * 100개

```python
### 위의 '반도체'에 관련된 100개 주변 단어에 대해서 '급등' Positive 단어 10개씩 추출해서 res_psng 에 저장
def psng(voca):
  words = model.wv.most_similar(voca, topn=100)
  res_psng = []
  for word in words:
    a = []
    w = word[0] ### w2v 결과로 나온 단어 하나
    try:
      ### positive 급등 관련 단어
      w2 = model.wv.most_similar(positive=[w,'급등'], topn=12)
    
    except:
      continue
    
    ### b에 단어, positive, 카테고리 저장
    res_psng.append((w, w2))

  return res_psng
```

#### 종목 뽑기 함수 (count_stock)
- 주변 단어와 관련된 단어들에서 종목명을 찾아 가장 높은 빈도를 찾기 위한 함수
- **input** : (주변단어, ['종목과 관련된 단어들']) * 100개
- **logic**
1. 주변 단어를 따로 리스트에 저장
2. 종목과 관련된 단어들 중에서 실제로 종목명에 속하는 것들을 리스트에 저장
3. 저장된 종목명 리스트에서 빈도 가장 높은 종목 하나 추출 (most_frequent 함수로)
- **output** : 연관 종목 1개, 주변 단어 n개

```python
### 종목 리스트 불러와서 리스트 형태로 저장
import pandas as pd
stock = pd.read_csv("/content/drive/MyDrive/ForColab/stock.csv")
stock_name = stock['Name'].to_list()
names=''
for name in stock_name:
  names += name+'\n'

names_lst = names.split('\n')
names_lst[:10]
```

```python
### W2V 결과에서 종목 이름 추출하여 가장 많이 나온 종목 출력
def most_frequent(data):
    return max(data, key=data.count)

def count_stock(data):
  stock_names = []
  around_words = []
  for i in range(len(data)):
    ### 해당 단어와 관련된 주변 단어 따로 저장
    around_word = data[i][0]
    around_words.append(around_word)
    
    for j in range(len(data[i])):
      for w in data[i][j]:
        if w[0] in names_lst:
          stock_names.append(w[0])
  stock = most_frequent(stock_names)
  return stock, around_words
```

```python
### 반도체 연관단어 10개에 대해서 다시 워드투백한 결과에서 종목과 주변 단어 뽑기
a = psng("반도체")
stock_name, around_words = count_stock(a)
print(stock_name,'\n',around_words[:5])
```
